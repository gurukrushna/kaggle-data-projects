```python
"""grouping all small no books languages to other languages """

pie_value = df['language_code'].value_counts()[df['language_code'].value_counts() > 10]

pie_value["other_languages"] = sum(df['language_code'].value_counts()[df['language_code'].value_counts() < 10])
pie_value
list(pie_value)
```

```python
labels = pie_value.index
plt.style.use('fivethirtyeight')

explode = [0]*len(pie_value)


"""exploding those with small values"""
for i,no in enumerate(list(pie_value)):
    if list(pie_value)[i]<100:
        explode[i] = 0.2
    
       
        
plt.pie(pie_value ,labels = labels ,explode = explode ,autopct='%1.1f%%' ,wedgeprops={'edgecolor':'black'})
#plt.legend(labels, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.tight_layout()
plt.show()
```
